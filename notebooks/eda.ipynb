```python
import os
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/iris.csv')
```

```python
df.shape
```

```python
sns.histplot(data=df, x="PetalWidthCm",hue="Species", bins=20, kde=True, element="step") 
plt.title('Distribucin del Ancho del Ptalo (cm) por Especie')
```

```python
sns.histplot(data=df, x="SepalWidthCm",hue="Species", bins=20, kde=True, element="step") 
plt.title('Distribucin del Ancho del Spalo (cm) por Especie')
```

```python
sns.histplot(data=df, x="SepalLengthCm",hue="Species", bins=20, kde=True, element="step") 
plt.title('Distribucin del Largo del Spalo (cm) por Especie')
```

```python

sns.histplot(data=df, x="PetalLengthCm", hue="Species", bins=20, kde=True, element="step")
plt.title('Distribucin del Largo del Ptalo (cm) por Especie')
```

```python
sns.pairplot(
    df,
    hue="Species",
    palette="viridis",
    corner=True,
    diag_kind="kde",       # Cambia histogramas por curvas de densidad
    plot_kws={"alpha": 0.8, "s": 20},  # Transparencia y tamao de puntos
    height=2               # Altura de cada subgrfico
)
plt.suptitle("Relaciones entre caractersticas morfolgicas de Iris", y=1.02)
```
